```python
import matplotlib.pyplot as plt
plt.figure('listas', figsize=(10,6), dpi=100)
plt.xlabel('valoresxxxxx')
plt.ylabel('valoresyyyy')
plt.figure('meses', dpi=100)
plt.xlabel('valoresxxxxx')
plt.ylabel('valoresyyyy')

lista1 = [11,2,3,5,6,23,55,12]
lista2 = [1,2,8,3,5,25,1]
lista3 = [10,5,3,5,23,23,45]

plt.figure('listas')

plt.plot(lista1)
plt.plot(lista2)
plt.figure('meses')

plt.plot(lista1, label='Enero')
plt.plot(lista2, label='Febrero')
plt.plot(lista3, label='Marzo')
plt.legend(loc='upper left')
```

```python
import numpy as np
import matplotlib.pyplot as plt
T = 2
x=np.linspace(1,10,1000)
y=np.sin(2*np.pi*x/T)
z=np.cos(2*np.pi*x/T)

plt.figure('Ejemplo de subplots', facecolor='yellow', edgecolor='blue', frameon=True)

plt.subplot(2,2,1)
plt.plot(x,y,'r')

plt.subplot(2,2,2)
plt.plot(x,z,'b')

plt.subplot(2,2,3)
plt.plot(x,y,'g')

plt.subplot(2,2,4)
plt.plot(x,z,'k')

plt.show()
```
