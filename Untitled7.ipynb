```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.mixture import GaussianMixture
import numpy as np

sns.set_context("talk", font_scale=1.5)

#usaremos la funcion de make_blobs de sklearn.datasets para crear un conjunto de datos simulado con 4 clusteres diferentes. El argumento centers=4 especifica 4 clusteres. Tambien especificamos cmo de apretado debe ser el clster usando cluster_std.

X,y = make_blobs(n_samples = 500, centers=4, cluster_std=2, random_state=2021)

#la funcion make_blobs nos da los datos simulados como una matriz numpy y las etiquetas como vector.Almacenemos los datos como dataframe de Pandas.

data = pd.DataFrame(X)

data.columns = ["X1","X2"]
data["cluster"]=y
data.head()

#con la funcion GaussianMix() de scikit learn, podemos ajustar nuestros datos a los modelos de mezcla. Uno de los parmetros clave a usar al ajustar el modelo de mezcla gaussiana es el nmero de clsteres en el conjunto de datos.

#Para este ejemplo, construyamos el modelo de mezcla gaussiana con 3 clusteres. Dado que simulamos los datos con 4 clusteres, sabemos que es incorrecto, pero sigamos adelante y ajustemos los datos con el modelo de mezcla gaussiana.

gmm = GaussianMixture(3,covariance_type='full', random_state=0).fit(data[["X1","X2"]])

#para los clusteres identificados, podemos obtener la ubicacin de los medios utilizando el mtodo "means_" en GaussianMixture.

gmm.means_


#usando la funcin predict(), tambin podemos predecir las etiquetas para los puntos de datos. En este ejemplo, obtenemmos los valores predichos para los datos de entrada.

labels = gmm.predict(data[["X1","X2"]])

#agregemos las etiquetas predichas a nuestro marco de datos.

data["predicted_cluster"] = labels

#Finalmente visualizamos los datos coloreando los puntos de datos con etiquetas predichas.

plt.figure(figsize=(9,7))

sns.scatterplot(data=data, x="X1", y="X2", hue="predicted_cluster", palette = ["red", "blue", "green"])

plt.savefig("fitting_Gaussian_Mixture_models_with_3_components_scikit_learn_Python.png", format='png', dpi=150)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.mixture import GaussianMixture
import numpy as np

sns.set_context("talk", font_scale=1.5)

#usaremos la funcion de make_blobs de sklearn.datasets para crear un conjunto de datos simulado con 4 clusteres diferentes. El argumento centers=4 especifica 4 clusteres. Tambien especificamos cmo de apretado debe ser el clster usando cluster_std.

X,y = make_blobs(n_samples = 500, centers=4, cluster_std=2, random_state=2021)

#la funcion make_blobs nos da los datos simulados como una matriz numpy y las etiquetas como vector.Almacenemos los datos como dataframe de Pandas.

data = pd.DataFrame(X)

data.columns = ["X1","X2"]
data["cluster"]=y
data.head()

#con la funcion GaussianMix() de scikit learn, podemos ajustar nuestros datos a los modelos de mezcla. Uno de los parmetros clave a usar al ajustar el modelo de mezcla gaussiana es el nmero de clsteres en el conjunto de datos.

#Para este ejemplo, construyamos el modelo de mezcla gaussiana con 3 clusteres. Dado que simulamos los datos con 4 clusteres, sabemos que es incorrecto, pero sigamos adelante y ajustemos los datos con el modelo de mezcla gaussiana.

gmm = GaussianMixture(3,covariance_type='full', random_state=0).fit(data[["X1","X2"]])

#para los clusteres identificados, podemos obtener la ubicacin de los medios utilizando el mtodo "means_" en GaussianMixture.

gmm.means_


#usando la funcin predict(), tambin podemos predecir las etiquetas para los puntos de datos. En este ejemplo, obtenemmos los valores predichos para los datos de entrada.

labels = gmm.predict(data[["X1","X2"]])

#agregemos las etiquetas predichas a nuestro marco de datos.

data["predicted_cluster"] = labels

#Finalmente visualizamos los datos coloreando los puntos de datos con etiquetas predichas.

plt.figure(figsize=(9,7))

sns.scatterplot(data=data, x="X1", y="X2", hue="predicted_cluster", palette = ["red", "blue", "green"])

plt.savefig("fitting_Gaussian_Mixture_models_with_3_components_scikit_learn_Python.png", format='png', dpi=150)

#podemos ver claramente que al ajustar el modelo con 3 clusteres es incorrecto. El modelo ha agrupado dos clsteres en uno.

#A menudo el mayor desafo es que nos sabremos los grupos de nmeros en el conjunto de datos. Necesitamos identificar correctamente el nmero de grupos. Una de las formas en que podemos hacerlo es ajustar el modelo de mezcla gaussiana con un nmero mltiple de grupos, digamos que van de 1 a 20.
#y luego hacemos una comparacin de modelos para encontrar que modelo se ajusta primero a los datos. Por ejemplo, es un modelo de mezcla gaussiana con 4 clusteres que encajan mmejor o un modelo con 3 clsteres que encajen mejor. Despus podemos seleccionar el mejor modelo con cierto nmero de clsteres que se ajuste a los datos.

#Las puntuaciones AIC o BIC se usan habitualmente para comparar modelos y seleccionar el mejor modelo que se ajuste a los datos. Una sola de las puntuaciones es lo suficientemente buena como para hacer una comparacin de modelos. Sin embargo vamos a calcular ambas puntuaciones, slo para ver su comportamiento.

#ajustamos los datos con el modelo de mezcla gaussiana con diferentes nmeros de grupos.

n_componentes = np.arange(1,21)

models = [GaussianMixture(n, covariance_type='full', random_state=0).fit(X) for n in n_componentes]
models[0:5]
[GaussianMixture(random_state=0), GaussianMixture(n_componentes=2, random_state=0), GaussianMixture(n_componentes=3, random_state=0), GaussianMixture(n_componentes=4, random_state=0), GaussianMixture(n_componentes=5, random_state=0)]

#podemos clacular facilmente los puntajes AIC / BIC con scikit_learn. Aqui usamos para uno de los modelos y calculamos las puntuaciones BIC y AIC

models[0].bic(X)
models[0].aic(X)
```
