```python
import matplotlib.pyplot as plt
```

```python
plt.figure('listas', figsize=(6,4), dpi=100)
plt.figure('meses', dpi=50)
lista1 = [11,2,3,5,6,23,55,12]
lista2 = [1,2,8,3,5,25,1]
lista3 = [10,5,3,5,23,23,45]
plt.figure('listas')
plt.plot(lista1)
plt.plot(lista2)
```
